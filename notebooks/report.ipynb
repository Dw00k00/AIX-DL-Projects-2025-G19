# I. Proposal

## Motivation
최근 유튜브·틱톡·커버 영상 플랫폼이 성장하면서,
“내가 부른 노래를 특정 가수의 음색으로 변환하고 싶다”는 수요가 크게 늘었다.
하지만 기존의 보컬 믹싱·튜닝 작업은 전문 장비와 기술이 필요하며
일반 사용자들이 직접 접근하기 어렵다.

딥러닝 기반 Singing Voice Conversion(SVC) 기술은

멜로디와 리듬은 그대로 유지하면서

음색(timbre)만 특정 가수의 스타일로 바꾸는 기술이다.

이 기술을 구현해보면
오디오 처리, 딥러닝 모델, self-supervised speech representation 등
현대 음성 AI가 어떻게 동작하는지 실제로 경험할 수 있다.

특히 이번 프로젝트의 목적은
복잡한 모델을 직접 구현하는 것이 아니라,
실제 오픈소스 SVC 모델을 활용해 전체 파이프라인을 이해하고 분석하는 것이다.

이를 통해 SVC 모델 구조 이해, 데이터 전처리와 특징 추출 과정 이해, 모델 결과 분석 능력, 음성 AI에 대한 실전 감각을 얻을 수 있다는 점이 큰 동기이다.

## What do you want to see at the end?
본 프로젝트의 최종 목표는 사용자가 부른 노래를 입력하면,
멜로디·가사·리듬은 그대로 유지한 채 특정 가수의 음색으로 자연스럽게 변환하는
완전한 Singing Voice Conversion(SVC) 파이프라인을 구축하는 것이다.

최종 산출물은 실제 보컬 입력에 대해

타깃 가수의 목소리 스타일로 변환된 오디오 결과물,

그 변환 과정에서의 음색 변화, pitch 보존 정도 등
모델의 동작 능력을 직접 확인할 수 있는 예시 출력들로 구성된다.

즉, 사용자가 업로드한 음성을
“특정 가수 스타일의 보컬로 자연스럽게 바꿔주는”
엔드 투 엔드 음성 스타일 전환 시스템의 구현 및 성능 시연이
본 프로젝트의 최종 목표이다.
